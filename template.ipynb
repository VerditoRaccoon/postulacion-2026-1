## Código solución (NO MODIFICAR)

```python
def conteo_votos(candidatos, locales_de_votacion):
    resultados = []

    for local in locales_de_votacion:
        votos_c1 = 0
        votos_c2 = 0
        votos_nulos = 0

        for voto in local:
            if voto not in candidatos:
                votos_nulos += 1
            elif voto == candidatos[0]:
                votos_c1 += 1
            elif voto == candidatos[1]:
                votos_c2 += 1

        if votos_nulos > votos_c1 and votos_nulos > votos_c2:
            resultados.append("Reeleccion")
        elif votos_c1 > votos_c2:
            resultados.append(candidatos[0])
        elif votos_c2 > votos_c1:
            resultados.append(candidatos[1])
        else:
            resultados.append("Reeleccion")

    return resultados
```
